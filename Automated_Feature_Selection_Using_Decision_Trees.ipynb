```python
import pandas as pd
df = pd.read_csv(r"D:\Automate_Feature_Selection_Using_Python\dynamic_pricing.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_tools = ["Location_Category","Customer_Loyalty_Status","Time_of_Booking","Vehicle_Type"]
```

```python
label_encoders = {}
for col in categorical_tools:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

Training a decision tree for feature selection

```python
X = df.drop(columns=["Historical_Cost_of_Ride"])
X
```

```python
Y = df['Historical_Cost_of_Ride']
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=42)
model.fit(X, Y)

feature_importances = pd.DataFrame({"feature":X.columns,"importance":model.feature_importances_})
feature_importances = feature_importances.sort_values(by='importance', ascending=False)
```

```python
feature_importances
```

Automate Feature Selection

```python
# define threshold
threshold = 0.01  # keep features with importance > 1%
selected_features = feature_importances[feature_importances['importance'] > threshold]['feature'].tolist()

# filter dataset
# filter dataset
X_selected = X[selected_features]
```
